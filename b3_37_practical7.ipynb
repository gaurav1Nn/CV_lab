```python
#SIFT SURF HARRIS CANNY -- EDGE & CORNER -- KEYWORDS
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
%matplotlib inline
image= cv2.imread('tree.jpg')

image_copy=np.copy(image)
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray=cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
grap=np.float32(gray)

#detect corners
dst=cv2.cornerHarris(gray,3,3,0.04)

dst=cv2.dilate(dst,None)

plt.imshow(dst,cmap='gray')
```

```python
thresh=0.1*dst.max()

corner_image=np.copy(image_copy)

for j in range(0,dst.shape[0]):
    for i in range(0,dst.shape[1]):
        if(dst[j,i]>thresh):
            cv2.circle(corner_image,(i,j),1,(0,255,0),1)
plt.imshow(corner_image)
```

```python
image = cv2.imread('tree.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

#Applying SIFT -- detect and describe keypoints in image, making it useful for feature matching,
#object recognition and image stitiching.
#keypoints-- brightness,position, blob, how strong the feature is
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(image, None)
sift_image = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

#Apply SURF
orb = cv2.ORB_create()
#detect keypoints and compute descriptors
keypoints, descriptors = orb.detectAndCompute(image, None)
#draw keypoints
surf_image = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(sift_image, cmap='gray')
axes[0].set_title('SIFT')
axes[0].axis('off')

axes[1].imshow(surf_image, cmap='gray')
axes[1].set_title('SURF')
axes[1].axis('off')
plt.show()
```

```python
image = cv2.imread('tree.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#apply canny edge detection
edges = cv2.Canny(gray, 50, 150) #thresholds: 50(low), 150(high)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
